```python
import matplotlib.pyplot as plt
import json
```

```python
def plot(filename, modelname):
    outfilename = filename.split('.')[0]

    with open(filename) as f:
        data = json.load(f)
    xs = [i for i in range(len(data['trainacc']))]
    fig = plt.figure(figsize=(10,2.5))
    plt.title(f'{modelname} Loss per Epoch')
    plt.plot(xs, data['trainloss'], '.-', label='Train Loss')
    plt.plot(xs, data['devloss'], '.-', label='Dev Loss')
    plt.legend()
    fig.savefig(f'{outfilename}_loss.png', dpi=200)

    xs = [i for i in range(len(data['trainacc']))]
    fig = plt.figure(figsize=(10,2.5))
    plt.title(f'{modelname} Accuracy per Epoch')
    plt.plot(xs, data['trainacc'], '.-', label='Train Accuracy')
    plt.plot(xs, data['devacc'], '.-', label='Dev Accuracy')
    plt.legend()
    fig.savefig(f'{outfilename}_acc.png', dpi=200)
```

```python
plot('cnn_hist_random.json', 'CNN Random Split - ')
```

```python
plot('cnn_hist_subject.json', 'CNN Subject Split - ')
```

```python
cardinal = '#980000'
cal = '#003262'
```

```python
with open('cnn_hist_random.json') as f:
    d = json.load(f)
    cnn_rand_train = d['trainacc']
    cnn_rand_dev = d['devacc']
with open('cnn_hist_subject.json') as f:
    d = json.load(f)
    cnn_subj_train = d['trainacc']
    cnn_subj_dev = d['devacc']
```

```python
xs = [i for i in range(len(cnn_subj_train))]
fig = plt.figure(figsize=(15,4))
plt.title('CNN Accuracy over Epoch')
plt.plot(xs, cnn_rand_train, '-', color=cardinal, label='train: random split')
plt.plot(xs, cnn_rand_dev, '--', color=cardinal, label='dev: random split')
plt.plot(xs, cnn_subj_train, '-', color=cal, label='train: subject split')
plt.plot(xs, cnn_subj_dev, '--', color=cal, label='dev: subject split')
plt.legend()
```

```python
fig.savefig('cnn_acc.png', dpi=200)
```
