```python
import matplotlib.pyplot as plt
import json
```

```python
def plot(filename, modelname):
    outfilename = filename.split('.')[0]

    with open(filename) as f:
        data = json.load(f)
    xs = [i for i in range(len(data['trainacc']))]
    fig = plt.figure(figsize=(10,2.5))
    plt.title(f'{modelname} Loss per Epoch')
    plt.plot(xs, data['trainloss'], '.-', label='Train Loss')
    plt.plot(xs, data['devloss'], '.-', label='Dev Loss')
    plt.legend()
    fig.savefig(f'{outfilename}_loss.png', dpi=200)

    xs = [i for i in range(len(data['trainacc']))]
    fig = plt.figure(figsize=(10,2.5))
    plt.title(f'{modelname} Accuracy per Epoch')
    plt.plot(xs, data['trainacc'], '.-', label='Train Accuracy')
    plt.plot(xs, data['devacc'], '.-', label='Dev Accuracy')
    plt.legend()
    fig.savefig(f'{outfilename}_acc.png', dpi=200)
```

```python
plot('cnn_hist_random.json', 'CNN Random Split - ')
```

```python
plot('cnn_hist_subject.json', 'CNN Subject Split - ')
```
